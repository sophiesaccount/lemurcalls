```python
import json
import librosa
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from IPython.display import Audio, display

# === Parameter ===
filter_class = None  # z.B. "vocal" oder None fr alle Klassen
show_only_false_positives = True  # True: nur False Positives anzeigen, False: alle anzeigen
min_max_amp_labels = 0.05         # z.B. 0.05 oder None (fr Labels)
min_max_amp_predictions = 0.05    # z.B. 0.05 oder None (fr Predictions/False Positives)

# === Lade Hauptlabels ===
wav_path = r"C:/Users/sophi/Documents/Masterarbeit/Results/pred/(2019_03_15-12_02_11)_CSWMUW240241_0000_first.wav"
#json_path = r"C:/Users/sophi/Documents/Masterarbeit/Results/pred/(2019_03_15-12_02_11)_CSWMUW240241_0000_first_pred.json"
json_path = r"C:/Users/sophi/.vscode/Lemur/lemurvocalizations24-25dierks/amplitude_filter/filtered_preds.json"
audio, sr = librosa.load(wav_path, sr=None)
with open(json_path, 'r') as f:
    labels = json.load(f)
onsets = np.array(labels['onset'])
offsets = np.array(labels['offset'])
clusters = labels['cluster']

# === Lade Referenzlabels (Ground Truth) ===
# Setze auf None, wenn du keine Referenzlabels hast!
#json_path_ref = r"C:\Users\sophi\Documents\Masterarbeit\Results\Raven\labels_yesno.json"
json_path_ref = r"C:/Users/sophi/.vscode/Lemur/lemurvocalizations24-25dierks/amplitude_filter/filtered_labels.json"
labels_ref = None
try:
    with open(json_path_ref, 'r') as f:
        labels_ref = json.load(f)
except FileNotFoundError:
    labels_ref = None

def is_false_positive(onset, offset, cluster, labels_ref, amp, min_max_amp_predictions, tol=0.2):
    """Gibt True zurck, wenn das Segment ein False Positive ist (kein Match im Referenzset) UND amplitude >= threshold."""
    if labels_ref is None:
        return False  # Wenn kein Referenzset, dann keine FP-Logik
    if min_max_amp_predictions is not None and amp < min_max_amp_predictions:
        return False  # Prediction zu leise
    ref_onsets = np.array(labels_ref['onset'])
    ref_offsets = np.array(labels_ref['offset'])
    ref_clusters = np.array(labels_ref['cluster'])
    for ro, rf, rc in zip(ref_onsets, ref_offsets, ref_clusters):
        # berlappung berechnen
        intersection = max(0, min(offset, rf) - max(onset, ro))
        union = max(offset, rf) - min(onset, ro)
        overlap_ratio = intersection / union if union > 0 else 0
        if overlap_ratio > tol and str(cluster) == str(rc):
            return False  # Es gibt ein Match
    return True  # Kein Match gefunden  False Positive

# === Plot fr max. 5 Segmente ===
count = 0
for i in range(len(onsets)):
    cluster = clusters[i]
    if filter_class is not None and cluster != filter_class:
        continue

    onset = onsets[i]
    offset = offsets[i]
    audio_segment = audio[int(onset * sr):int(offset * sr)]

    # Amplitudenwerte berechnen
    max_amp = np.max(np.abs(audio_segment))
    mean_amp = np.mean(np.abs(audio_segment))

    # Label-Threshold
    if min_max_amp_labels is not None and max_amp < min_max_amp_labels:
        continue

    # False Positive Filter (inkl. Prediction-Threshold)
    if show_only_false_positives and labels_ref is not None:
        if not is_false_positive(onset, offset, cluster, labels_ref, max_amp, min_max_amp_predictions):
            continue

    print(f"Label {i+1}: {cluster} ({onset:.2f}s - {offset:.2f}s)")
    print(f"  Maximale Amplitude: {max_amp:.4f}")
    print(f"  Durchschnittliche Amplitude: {mean_amp:.4f}\n")

    # Spektrogramm-Parameter
    n_fft = 512 if sr <= 32000 else 1024
    hop_length = 128
    n_mels = 80
    fmax = sr / 2

    # Mel-Spektrogramm
    S = librosa.feature.melspectrogram(
        y=audio_segment, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, fmax=fmax
    )
    S_dB = librosa.power_to_db(S, ref=np.max)

    # Colorize
    cmap = cm.get_cmap('magma')
    S_norm = (S_dB - S_dB.min()) / (S_dB.max() - S_dB.min() + 1e-12)
    S_color = cmap(S_norm)[:, :, :3]
    S_color = np.flipud(S_color)

    # Plot
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.imshow(S_color, aspect='auto', origin='upper')

    # X-Achse
    n_frames = S.shape[1]
    duration = len(audio_segment) / sr
    t = np.linspace(0, duration, n_frames)
    xticks = np.linspace(0, n_frames-1, 5)
    ax.set_xticks(xticks)
    ax.set_xticklabels([f'{t[int(x)]:.2f}' for x in xticks])
    ax.set_xlabel('Time (s)')

    # Y-Achse
    mel_bin_freqs = np.linspace(sr / 2, 0, n_mels)
    tick_freqs_khz = np.arange(0, int(sr / 2 / 1000) + 1, 1)
    tick_positions = [np.argmin(np.abs(mel_bin_freqs - f * 1000)) for f in tick_freqs_khz]
    tick_labels = [f'{f}' for f in tick_freqs_khz]
    ax.set_yticks(tick_positions)
    ax.set_yticklabels(tick_labels)
    ax.set_ylabel('Frequency (kHz)')

    ax.set_title(f'Spectrogram for Label {i+1}: {cluster}\n({onset:.2f}s - {offset:.2f}s)')
    plt.tight_layout()
    plt.show()

    # Audio player
    display(Audio(data=audio_segment, rate=sr))

    count += 1
    if count >= 5:
        break
```

```python
def evaluate_detection_metrics(labels, predictions, overlap_tolerance=0.2):
    """
    Berechnet TP, FP, FN und F1 fr gegebene Labels und Predictions.
    labels, predictions: Dicts mit keys 'onset', 'offset', 'cluster' (Listen)
    overlap_tolerance: Mindestberlappung (0...1), damit ein Match als TP zhlt
    Rckgabe: dict mit keys 'tp', 'fp', 'fn', 'f1'
    """
    import numpy as np

    label_onsets = np.array(labels['onset'])
    label_offsets = np.array(labels['offset'])
    label_clusters = np.array(labels['cluster'])

    pred_onsets = np.array(predictions['onset'])
    pred_offsets = np.array(predictions['offset'])
    pred_clusters = np.array(predictions['cluster'])

    matched_labels = set()
    matched_preds = set()

    # True Positives: Prediction und Label matchen in Zeit und Klasse
    for p_idx, (po, pf, pc) in enumerate(zip(pred_onsets, pred_offsets, pred_clusters)):
        for l_idx, (lo, lf, lc) in enumerate(zip(label_onsets, label_offsets, label_clusters)):
            if l_idx in matched_labels:
                continue
            # berlappung berechnen
            intersection = max(0, min(pf, lf) - max(po, lo))
            union = max(pf, lf) - min(po, lo)
            overlap_ratio = intersection / union if union > 0 else 0
            if overlap_ratio > overlap_tolerance and str(pc) == str(lc):
                matched_labels.add(l_idx)
                matched_preds.add(p_idx)
                break

    tp = len(matched_labels)
    fp = len(pred_onsets) - len(matched_preds)
    fn = len(label_onsets) - len(matched_labels)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall    = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1        = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    return {'tp': tp, 'fp': fp, 'fn': fn, 'f1': f1}
```

```python
metrics = evaluate_detection_metrics(labels, labels_ref, overlap_tolerance=0.2)
print("True Positives:", metrics['tp'])
print("False Positives:", metrics['fp'])
print("False Negatives:", metrics['fn'])
print("F1-Score:", metrics['f1'])
```
